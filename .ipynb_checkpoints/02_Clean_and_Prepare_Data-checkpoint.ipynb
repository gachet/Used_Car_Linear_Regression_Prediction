## Data Cleaning and Preping

### Introduction

The following was done to prepare the data to be transformed into a data frame and later start the EDA and Feature engineering process

```python
import json
import re
```

```python
with open("scrape_json/cars_data_full.json", "r") as read_file:
    data1 = json.load(read_file)
```

```python
with open("scrape_json/cars_data_2100.json", "r") as read_file:
    data = json.load(read_file)
```

```python
all_website_makes = ['Acura', 'Alfa Romeo', 'Aston Martin', 'Audi', 'Bentley', 'BMW', 'Buick', 'Cadillac', 'Chevrolet',
         'Chrysler', 'Dodge', 'Ferrari', 'FIAT', 'Ford', 'Genesis', 'GMC', 'Honda', 'Hyundai', 'INFINITI', 'Jaguar',
         'Jeep', 'Kia', 'Lamborghini', 'Land Rover', 'Lexus', 'Lincoln', 'Lotus', 'Maserati', 'Mazda', 'McLaren',
         'Mercedes-Benz', 'MINI', 'Mitsubishi', 'Nissan', 'Porsche', 'RAM', 'Rolls-Royce', 'Scion', 'smart', 'Subaru',
         'Tesla', 'Toyota', 'Volkswagen', 'Volvo']
```

```python
# Regex was used to extract the Make, Model and Year from the 'Brand' line in the dictionary
# As an example:
# 'brand' : '2016 Nissan Murano S'
# We add three more keys with the following values
# 'year': 2016 (as an integer)
# 'make': Nissan
# 'model': Murano
```

```python
for t in data:
    if re.findall('(?<=20\d\d )(Alfa Romeo|Land Rover|Aston Martin|[a-zA-Z??-]*)(?= )', t['brand']):
        t['make'] = re.findall('(?<=20\d\d )(Alfa Romeo|Land Rover|Aston Martin|[a-zA-Z??-]*)(?= )', t['brand'])[0]
    else:
        t['make'] = 'Unknown'
    t['year'] = int(re.findall('201.', t['brand'])[0])
    if re.findall('(?:Acura |Alfa Romeo |Aston Martin |Audi |Bentley |BMW |Buick |Cadillac |Chevrolet |Chrysler |Dodge |Ferrari |FIAT |Ford |Genesis |GMC |Honda |Hyundai |INFINITI |Jaguar |Jeep |Kia |Lamborghini |Land Rover |Lexus |Lincoln |Lotus |Maserati |Mazda |McLaren |Mercedes-Benz |MINI |Mitsubishi |Nissan |Porsche |RAM |Rolls-Royce |Scion |smart |Subaru |Tesla |Toyota |Volkswagen |Volvo )([a-zA-Z0-9]*)(?= )', t['brand']):
        t['model'] = re.findall('(?:Acura |Alfa Romeo |Aston Martin |Audi |Bentley |BMW |Buick |Cadillac |Chevrolet |Chrysler |Dodge |Ferrari |FIAT |Ford |Genesis |GMC |Honda |Hyundai |INFINITI |Jaguar |Jeep |Kia |Lamborghini |Land Rover |Lexus |Lincoln |Lotus |Maserati |Mazda |McLaren |Mercedes-Benz |MINI |Mitsubishi |Nissan |Porsche |RAM |Rolls-Royce |Scion |smart |Subaru |Tesla |Toyota |Volkswagen |Volvo )([a-zA-Z0-9]*)(?= )', t['brand'])[0]
    else:
        t['model'] = 'Unknown'
```
